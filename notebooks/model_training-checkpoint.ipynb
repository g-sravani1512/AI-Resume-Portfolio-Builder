```python
import pandas as pd

# Load the raw dataset
df_raw = pd.read_csv("resume_dataset.csv")

# Check raw data
df_raw.head()
```

```python
# The entire row is currently in one column
col = df_raw.columns[0]

# Split into 5 columns using comma
df = df_raw[col].str.split(',', n=4, expand=True)

# Rename columns properly
df.columns = ['resume_id', 'category', 'resume_text', 'skills_list', 'experience_years']

# Show fixed dataset
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df['category'].value_counts()
```

```python
# Check unique categories
df['category'].unique()
```

```python
# Remove rows where category is numeric
df = df[~df['category'].str.replace('.', '', regex=False).str.isnumeric()]

# Remove rows where category is too long (likely resume text)
df = df[df['category'].str.len() < 40]

# Remove missing values
df = df.dropna(subset=['category'])
```

```python
df['category'].value_counts()
```

```python
import re

def clean_text(text):
    text = str(text).lower()                     # convert to lowercase
    text = re.sub(r'\d+', '', text)             # remove numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)     # remove punctuation/symbols
    text = re.sub(r'\s+', ' ', text)            # remove extra spaces
    return text

# Apply cleaning to resume_text
df['cleaned_resume'] = df['resume_text'].apply(clean_text)

# Check result
df[['resume_text', 'cleaned_resume']].head()
```

```python
# Input features (what the model reads)
X = df['cleaned_resume']

# Target labels (what the model predicts)
y = df['category']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english', max_features=5000)

X_tfidf = tfidf.fit_transform(X)

X_tfidf.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf, y, test_size=0.2, random_state=42
)

X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression

# Create the model
model = LogisticRegression(max_iter=1000)

# Train the model
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Predict on test data
y_pred = model.predict(X_test)

# Check accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy:", accuracy)
```

```python
from sklearn.svm import LinearSVC

# Create SVM model
svm_model = LinearSVC()

# Train the model
svm_model.fit(X_train, y_train)

# Predict
y_pred_svm = svm_model.predict(X_test)

# Check accuracy
from sklearn.metrics import accuracy_score
svm_accuracy = accuracy_score(y_test, y_pred_svm)

print("SVM Model Accuracy:", svm_accuracy)
```

```python
# Count samples per category
category_counts = df['category'].value_counts()

# Keep categories with at least 20 samples
valid_categories = category_counts[category_counts >= 20].index

# Filter dataset
df_filtered = df[df['category'].isin(valid_categories)]

print(df_filtered['category'].value_counts())
```

```python
X = df_filtered['cleaned_resume']
y = df_filtered['category']

from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english', max_features=5000)
X_tfidf = tfidf.fit_transform(X)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

svm_model = LinearSVC()
svm_model.fit(X_train, y_train)

y_pred = svm_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print("Improved SVM Accuracy:", accuracy)
```

```python
from sklearn.svm import LinearSVC
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.1, 1, 5, 10]
}

svm = LinearSVC()

grid = GridSearchCV(svm, param_grid, cv=3, scoring='accuracy', n_jobs=-1)
grid.fit(X_train, y_train)

print("Best Parameter:", grid.best_params_)
print("Best Accuracy:", grid.best_score_)
```

```python
best_svm = grid.best_estimator_

y_pred_best = best_svm.predict(X_test)
from sklearn.metrics import accuracy_score
final_accuracy = accuracy_score(y_test, y_pred_best)

print("Final Tuned SVM Accuracy:", final_accuracy)
```

```python
for c in [0.1, 0.5, 1, 5, 10]:
    svm = LinearSVC(C=c)
    svm.fit(X_train, y_train)
    y_pred = svm.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"C={c}  Accuracy={acc}")
```

```python
from sklearn.svm import LinearSVC
import joblib

# Train final model with best C value
best_svm_model = LinearSVC(C=0.5)
best_svm_model.fit(X_train, y_train)

# Save the model and vectorizer
joblib.dump(best_svm_model, "resume_classifier_model.pkl")
joblib.dump(tfidf, "tfidf_vectorizer.pkl")

print("Best model and vectorizer saved successfully!")
```
